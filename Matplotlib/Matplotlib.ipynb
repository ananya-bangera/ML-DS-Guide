```python
# Basic Graph
x = sorted(np.random.randint(0,100,25))
y = sorted(np.random.randint(0,100,25))

# Resize your Graph
# plt.figure(figsize=(8,6),dpi=100)

# plt.plot(x,y,label='Random',color='r',linewidth=3,marker='*',markersize=10,markeredgecolor='b')
# Shorthand Notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'r-4',label='Random')

# Line 2
y1 = sorted(np.random.randint(0,100,25))
y1.reverse()
plt.plot(x[:10],y1[:10],label='Rev. Random')
plt.plot(x[9:],y1[9:],'b--4',label='Rev. Random Prediction')

# Axes Labels & Titles
plt.title('Increasing Function',fontdict={'fontsize':20})
plt.xlabel('X - Value')
plt.ylabel('Y - Value')

# X,Y bins/ticks size 
plt.xticks([i for i in range(0,101,10)])
plt.yticks([i for i in range(0,101,15)])

plt.legend()

# Save the plot 
# plt.savefig('mygraph.png',dpi=300)

# x,y
# Show plot
plt.show()
```

```python
# Bar Chart
labels = list('ABCD')
pattern = '/o*\\'
values = np.random.randint(10,100,4)

bar = plt.bar(labels,values)
for i in range(len(pattern)) :
  bar[i].set_hatch(pattern[i])

plt.show()
```

```python
fifa = pd.read_csv('/content/drive/MyDrive/ML Datasets/fifa_data.csv')

plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols = ['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 
        'NumOfProducts', 'IsActiveMember', 'EstimatedSalary', 'Exited']
churn = pd.read_csv("/content/drive/MyDrive/ML Datasets/Churn_Modelling.csv", usecols=cols)
churn.head()
```

```python
# 1. Number of customers in each country
plt.figure(figsize=(8,5))
plt.title('Number of Customers',fontsize=14)

x = churn['Geography'].value_counts()
plt.bar(x=x.index,height=x.values)
plt.show()
```

```python
# 2. Adjusting xticks and yticks
plt.figure(figsize=(8,5))
plt.title('Number of Customers',fontsize=14)

x = churn['Geography'].value_counts()
plt.bar(x=x.index,height=x.values)

plt.xticks(fontsize=12,rotation=45)
plt.yticks(ticks=np.arange(0,7000,1000),fontsize=12)

plt.show()
```

```python
# Changing the default figure size
print(plt.rcParams.get('figure.figsize'))
plt.rcParams['figure.figsize']=(8,5)
print(plt.rcParams.get('figure.figsize'))
```

```python
# 4. Creating a simple histogram
plt.hist(x=churn['Balance'])
plt.show()
```

```python
# 5. Customizing the histogram
plt.hist(x=churn['Balance'],bins=12,color='darkgrey',range=(25000,225000))
plt.title("Distribution on Balance (25000 - 225000)",fontsize=14)
plt.show()
```

```python
# 6. Creating a simple scatter plot
sample = churn.sample(n=200,random_state=42) 
plt.scatter(x=sample['CreditScore'],y=sample['Age'])
plt.show()
```

```python
# 7. Scatter plots with subplots
fig, ax = plt.subplots()
plt.title("France vs Germany", fontsize=14)
ax.scatter(x=sample[sample.Geography == 'France']['CreditScore'],
           y=sample[sample.Geography == 'France']['Age'])
ax.scatter(x=sample[sample.Geography == 'Germany']['CreditScore'],
           y=sample[sample.Geography == 'Germany']['Age'])
ax.legend(labels=['France','Germany'], loc='lower left', fontsize=12)
```

```python
# 8. Grid of subplots
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3)
```

```python
# 9. Rearranging and accessing the subplots
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(9,6), sharex=True)
fig.tight_layout(pad=2)
```

```python
# 10. Drawing the subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True,
figsize=(8,5))
countries = churn.Geography.value_counts()
products = churn.NumOfProducts.value_counts()
ax1.bar(x=countries.index, height=countries.values)
ax1.set_title("Countries", fontsize=12)
ax2.bar(x=products.index, height=products.values)
ax2.set_title("Number of Products", fontsize=12)
```

```python
# 11. Creating a 2-D histogram
plt.title("Credit Score vs Age", fontsize=15)
plt.hist2d(x=churn.CreditScore, y=churn.Age)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML Datasets/company_sales_data.csv')
df.head()
```

```python
# Exercise 1: Read Total profit of all months and show it using a line plot
plt.plot(df.month_number,df.total_profit)
```

```python
# Exercise 2: Get total profit of all months and show line plot with the following Style properties
# Generated line plot must include following Style properties: –
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3

plt.plot(df.month_number,df.total_profit,'r--o',linewidth=3,mec='k',
         label='Profit data of last year')
plt.title('Company Sales data of last year')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
```

```python
# Exercise 3: Read all product sales data and show it  using a multiline plot

plt.figure(figsize=(12,8))
plt.title("Sales data")
df.columns[1:-2].values
for i in df.columns[1:-2].values :
  plt.plot(df['month_number'],df[i],'-o',label=f'{i} Sales Data')

plt.legend(loc="upper left")
plt.show()
```

```python
# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

plt.figure(figsize=(12,8))
plt.scatter(df.month_number,df['toothpaste'],label="Tooth paste Sales data")

plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.grid(True,linewidth=1,linestyle='--')
plt.xticks([i for i in range(14)])
plt.legend()
plt.show()
```

```python
# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

plt.figure(figsize=(12,8))
plt.title('Facewash & Facecream sales data')
plt.bar(df.month_number-0.125,df['facecream'],label='Face Cream Sales Data',width=0.25)
plt.bar(df.month_number+0.125,df['facewash'],label='Face Wash Sales Data',width=0.25)
plt.xticks([i for i in range(14)])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend()
plt.grid(True,linewidth=1,linestyle='--')
plt.show()
```

```python
# Exercise 6: Read sales data of bathing soap of all months and show it using 
# a bar chart. Save this plot to your hard disk

plt.figure(figsize=(10,7))

plt.bar(df['month_number'],df['bathingsoap'])
plt.xticks([i for i in range(14)])
plt.title('bathingsoap sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.grid(True,linewidth=1,linestyle='--')
plt.show()
```

```python
# Exercise 7: Read the total profit of each month and show it using the
#  histogram to see the most common profit ranges


plt.figure(figsize=(10,7))
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(df['total_profit'],profit_range,range=(150000,350000),label="Profit data")
plt.title('Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend()
plt.show()
```

```python
# Exercise 8: Calculate total sale data for last year for each product and 
# show it using a Pie chart
plt.figure(figsize=(10,7))
plt.pie(df.iloc[:,1:-2].sum(),labels=df.columns.values[1:-2],autopct='%1.1f%%')

plt.legend(loc='lower right')
plt.show()
```

```python
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

f,a = plt.subplots(2,sharex=True)
a[0].plot(df['month_number'],df['bathingsoap'],'k-o')
a[0].set_title('Sales data of  a Bathingsoap')
a[1].plot(df['month_number'],df['facewash'],'r-o')
a[1].set_title('Sales data of  a facewash')

plt.xlabel('Month Number')
plt.ylabel('Sales1 Units in Number')
plt.show()
```

```python
# Exercise Question 10: Read all product sales data and show it using the stack plot
x = np.vstack([df[i] for i in df.columns[1:-2]])
plt.stackplot(df.iloc[:,0].values,x,labels=df.columns[1:-2])
plt.title('All product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.legend(loc='upper left')
plt.show()
```
