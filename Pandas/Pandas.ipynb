```python
# How to filter words that contain atleast 2 vowels from a series?
s = pd.Series(['Apple', 'Orange', 'Plan', 'Python', 'Money'])
vowels = list('aeiou')
from collections import Counter
s[s.map(lambda x: sum([Counter(x.lower())[i] for i in vowels]) >= 2)]
```

```python
# How to filter valid emails from a series?
import re
emails = pd.Series(['buying books at amazom.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com'])
pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'
print(emails[emails.map(lambda x:bool(re.match(pattern,x)))])
print(emails.str.findall(pattern,flags=re.IGNORECASE))
```

```python
# How to create a dataframe with rows as strides from a given series?
L = pd.Series(range(15))

def gen_strides(a, stride_len=5, window_len=5):
    n_strides = ((a.size-window_len)//stride_len) + 1
    return np.array([a[s:(s+window_len)] for s in np.arange(0, a.size, stride_len)[:n_strides]])

gen_strides(L, stride_len=2, window_len=4)
```

```python
#  How to find and cap outliers from a series or dataframe column?
# Replace all values of ser in the lower 5%ile and greater than 95%ile with respective
# 5th and 95th %ile value.
ser = pd.Series(np.logspace(-2, 2, 30))

x,y= 0.05,0.95
l,h=ser.quantile([x,y])
ser[ser < l] = l
ser[ser > h] = h
ser
```
